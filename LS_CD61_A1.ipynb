```python
arr1 = np.random.randint(1, 51, size=(5, 4))
print("Array:\n", arr1)
```

```python
anti_diag = [int(arr1[i, -i - 1]) for i in range(min(arr1.shape))]
print("Anti-diagonal:", anti_diag)
```

```python
row_max = np.max(arr1, axis=1)
print("Row-wise Max:", row_max)
```

```python
mean_val = np.mean(arr1)
less_eq_mean = arr1[arr1 <= mean_val]
print("Elements <= mean (", mean_val, "):\n", less_eq_mean)
```

```python
def numpy_boundary_traversal(matrix):
    rows, cols = matrix.shape
    boundary = []
    for j in range(cols):
        boundary.append(int(matrix[0, j]))
    for i in range(1, rows):
        boundary.append(int(matrix[i, cols - 1]))
    if rows > 1:
        for j in range(cols - 2, -1, -1):
            boundary.append(int(matrix[rows - 1, j]))
    if cols > 1:
        for i in range(rows - 2, 0, -1):
            boundary.append(int(matrix[i, 0]))

    return boundary


print("Boundary Traversal:", numpy_boundary_traversal(arr1))
```

```python
def numpy_alternate_sort(array):
    sorted_arr = np.sort(array)
    result = []
    i, j = 0, len(sorted_arr) - 1
    while i <= j:
        result.append(sorted_arr[i])
        if i != j:
            result.append(sorted_arr[j])
        i += 1
        j -= 1
    return np.array(result)

print("Alternate Sort:", numpy_alternate_sort(arr2))
```

```python
names = [f"Student{i}" for i in range(1, 11)]
subjects = ["Math", "Science", "History", "English", "Art"]
data = {
    "Name": names,
    "Subject": np.random.choice(subjects, size=10),
    "Score": np.random.randint(50, 101, size=10),
    "Grade": [""] * 10
}
df = pd.DataFrame(data)
df
```

```python
conditions = [
    (df['Score'] >= 90),
    (df['Score'] >= 80),
    (df['Score'] >= 70),
    (df['Score'] >= 60),
    (df['Score'] < 60)
]
grades = ['A', 'B', 'C', 'D', 'F']
df['Grade'] = np.select(conditions, grades,default='F')
df
```

```python
df_sorted = df.sort_values(by='Score', ascending=False)
print(df_sorted)
```

```python
print("Average Score per Subject:")
print(df.groupby('Subject')['Score'].mean())
```

```python
def pandas_filter_pass(dataframe):
    return dataframe[dataframe['Grade'].isin(['A', 'B'])]

print("Passed Students:")
print(pandas_filter_pass(df))
```

```python
def text_preprocess_vectorize(texts, vectorizer):
    return vectorizer.transform(texts)

sample_texts = ["This product is amazing!", "Worst thing I ever bought."]
sample_vectors = text_preprocess_vectorize(sample_texts, vectorizer)
predictions = model.predict(sample_vectors)
print("Predictions for sample texts:", list(zip(sample_texts, predictions)))
```
